Here, we can examine the results from the tabular experiments.

```python
import pandas as pd

model_type = "mlp"
start_drop = 1
end_drop = 99
dropcls_string = f"{start_drop}-{end_drop}" if start_drop != end_drop else str(start_drop)

experiment_path = f"../tabular/logs/purchase_{model_type}_qmia_summary_dropped_{dropcls_string}_seed_42.csv"
df = pd.read_csv(experiment_path)
# Round all numeric columns to 3 decimal points
numeric_columns = df.select_dtypes(include=['float64']).columns
for col in numeric_columns:
    if "tpr" in col:
        df[col] = df[col]*100
df[numeric_columns] = df[numeric_columns].round(3)
# Create new dataframe with metrics for unseen then seen classes
metrics = [
    ('unseen_qmia_auc', 'unseen_qmia_auc'),
    ('unseen_baseline_auc', 'unseen_baseline_auc'), 
    ('unseen_qmia_tpr_at_1_fpr', 'unseen_qmia_tpr_at_1_fpr'),
    ('unseen_baseline_tpr_at_1_fpr', 'unseen_baseline_tpr_at_1_fpr'),
    ('seen_qmia_auc', 'seen_qmia_auc'),
    ('seen_baseline_auc', 'seen_baseline_auc'),
    ('seen_qmia_tpr_at_1_fpr', 'seen_qmia_tpr_at_1_fpr'), 
    ('seen_baseline_tpr_at_1_fpr', 'seen_baseline_tpr_at_1_fpr'),
    ('overall_qmia_auc', 'overall_qmia_auc'),
    ('overall_baseline_auc', 'overall_baseline_auc'),
    ('overall_qmia_tpr_at_1_fpr', 'overall_qmia_tpr_at_1_fpr'), 
    ('overall_baseline_tpr_at_1_fpr', 'overall_baseline_tpr_at_1_fpr')
]

results = []
for metric_name, col_name in metrics:
    results.append({
        'metric': metric_name,
        'value': df[col_name].iloc[0]
    })

# Convert to dataframe
df = pd.DataFrame(results)

display(df)

[print(val) for val in df['value']]
```
